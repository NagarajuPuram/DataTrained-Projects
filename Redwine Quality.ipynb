```python
# create train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=125)
```

```python
# removing inbalance using smote 
from imblearn.over_sampling import SMOTE
sm=SMOTE()
over_smp=SMOTE(0.75)
x_train_rs,y_train_rs=over_smp.fit_resample(x_train,y_train)
```

```python
y_train
```

```python
y_train_rs
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
lr=LogisticRegression()
lr.fit(x_train_rs,y_train_rs)
predlr=lr.predict(x_test)
```

```python
print('accuracy:',accuracy_score(y_test,predlr)*100)
print('classification report:',classification_report(y_test,predlr))
print('confusion matrix:',confusion_matrix(y_test,predlr))
```

```python
# Decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train_rs,y_train_rs)
preddt=dt.predict(x_test)
```

```python
print('accuracy:',accuracy_score(y_test,preddt)*100)
print('classification report:',classification_report(y_test,preddt))
print('confusion matrix:',confusion_matrix(y_test,preddt))
```

```python
# Random forest classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train_rs,y_train_rs)
predrf=rf.predict(x_test)

print('accuracy:',accuracy_score(y_test,predrf)*100)
print('classification report:',classification_report(y_test,predrf))
print('confusion matrix:',confusion_matrix(y_test,predrf))
```

```python
# cross validation for overfitting
```

```python
from sklearn.model_selection import cross_val_score
# for logistic regression
cvr=cross_val_score(lr,x,y,cv=5)
print(cvr.mean())
```

```python
# for decision tree classifier
cvr=cross_val_score(dt,x,y,cv=5)
print(cvr.mean())
```

```python
# for random forest classifier
cvr=cross_val_score(rf,x,y,cv=5)
print(cvr.mean())
```
